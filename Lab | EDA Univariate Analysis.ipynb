# Part 1: Understanding Product Categories
Business Question: What are the most popular product categories on Amazon UK, and how do they compare in terms of listing frequency?

```python
#Generate frequency table for the product category
frequency_table = pd.crosstab(index = df['category'], columns = 'count')
frequency_table
```

```python
#Which are the top 5 most listed product categories?
frequency_table_sorted = frequency_table.sort_values(by = 'count', ascending = False)
sub_frequency_table = frequency_table_sorted.head(10)
sub_frequency_table
```

```python
#Display the distribution of products across different categories using a bar chart
sns.barplot(y=sub_frequency_table.index, x=sub_frequency_table["count"], palette="Set3")
```

```python
#For a subset of top categories, visualize their proportions using a pie chart.
sub_frequency_table['count'].plot.pie(autopct='%1.1f%%', startangle=45, colors=sns.color_palette("Set3"))
```

# Part 2: Delving into Product Pricing
Business Question: How are products priced on Amazon UK, and are there specific price points or ranges that are more common?

```python
#Measures of Centrality:
mean_price = df['price'].mean().round(2)
print('The mean price is:', mean_price)

median_price = df['price'].median()
print('The median price is:', median_price)

mode_price = df['price'].mode()[0]
print('The modal price is:', mode_price)

print(f'The average price point of products listed is {mean_price}, which is must higher than the mode, {mode_price}')
```

```python
#Measures of Dispersion:
var_price = df['price'].var()
print("The variance is:", var_price)

std_price = df['price'].std()
print('The standard deviation is:', std_price)

range_price = df['price'].max() - df['price'].min()
print('The range is:', range_price)

q1 = df['price'].quantile(0.25)
q3 = df['price'].quantile(0.75)
iqr = q3 - q1
print('The interquartile range is:', iqr)
print('')
print("The high standard deviation indicates that prices are more spread out from the mean.")
```

```python
#Plot a histogram to visualize the distribution of product prices
sns.histplot(df['price'], kde=True, bins=50, color="salmon")
```

```python
sns.boxplot(data = df['price'], color="salmon")
```

# Part 3: Unpacking Product Ratings
Business Question: How do customers rate products on Amazon UK, and are there any patterns or tendencies in the ratings?

```python
#Measures of Centrality:
mean_rating = df['stars'].mean().round(2)
print('The mean rating is:', mean_rating)

median_rating = df['stars'].median()
print('The median rating is:', median_rating)

mode_rating = df['stars'].mode()[0]
print('The modal rating is:', mode_rating)
```

```python
#Measures of Dispersion:
var_rating = df['stars'].var()
print("The variance is:", var_rating)

std_rating = df['stars'].std()
print('The standard deviation is:', std_rating)

range_rating = df['stars'].max() - df['stars'].min()
print('The range is:', range_rating)

q1 = df['stars'].quantile(0.25)
q3 = df['stars'].quantile(0.75)
iqr_rating = q3 - q1
print('The interquartile range is:', iqr_rating)
```

```python
#Shape of the Distribution:
skewness_rating = df['stars'].skew()
kurtosis_rating = df['stars'].kurtosis()

skewness_rating, kurtosis_rating
```
